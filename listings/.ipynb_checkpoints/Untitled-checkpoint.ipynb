```python
import numpy as np
import pandas as pd

from pandas import DataFrame, Series

dataset = pd.read_csv("austin_listings.csv", header = 0)
shared = []
entire = []
room = []
for row in dataset.itertuples():
    r_type = row.room_type
    if r_type == 'Shared room':
        room.append(0)
        shared.append(1)
        entire.append(0)
    elif r_type == 'Entire home/apt':
        room.append(0)
        shared.append(0)
        entire.append(1)
    elif r_type == 'Private room':
        room.append(1)
        shared.append(0)
        entire.append(0)
dataset['shared'] = Series(shared, index=dataset.index)
dataset['entire'] = Series(entire, index=dataset.index)
dataset['room'] = Series(room, index=dataset.index)
```
